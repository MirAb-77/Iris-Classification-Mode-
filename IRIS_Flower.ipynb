# The Iris Flower Data Set

The **Iris flower data set** or **Fisher's Iris data set** is a multivariate data set introduced by the British statistician, eugenicist, and biologist **Ronald Fisher** in his 1936 paper *The use of multiple measurements in taxonomic problems* as an example of linear discriminant analysis„Äê[1](https://en.wikipedia.org/wiki/Iris_flower_data_set#:~:text=The%20Iris%20flower%20data%20set,example%20of%20linear%20discriminant%20analysis.)„Äë. It is sometimes called **Anderson's Iris data set** because **Edgar Anderson** collected the data to quantify the morphologic variation of Iris flowers of three related species„Äê[2](https://en.wikipedia.org/wiki/Iris_flower_data_set#:~:text=The%20Iris%20flower%20data%20set,example%20of%20linear%20discriminant%20analysis.)„Äë.

Two of the three species were collected in the Gasp√© Peninsula "all from the same pasture, and picked on the same day and measured at the same time by the same person with the same apparatus"„Äê[3](https://en.wikipedia.org/wiki/Iris_flower_data_set#:~:text=The%20Iris%20flower%20data%20set,example%20of%20linear%20discriminant%20analysis.)„Äë. Fisher's paper was published in the journal, the **Annals of Eugenics**, creating controversy about the continued use of the Iris dataset for teaching statistical techniques today.

The data set consists of `150` records and `50` samples under 5 attributes from each of three species(Class) of Iris (`Iris setosa`, `Iris virginica` and `Iris versicolor`). Four features were measured from each sample: the `length` and the `width` of the `sepals` and `petals`, in centimeters(Petal Length, Petal Width, Sepal Length, Sepal width). Based on the combination of these four features, Fisher developed a linear discriminant model to distinguish the species from each other.You can download it [here](https://www.kaggle.com/datasets/arshid/iris-flower-dataset).

```python
def plot_confusion_matrix(cm, class_names):
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False,
                xticklabels=class_names, yticklabels=class_names)
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()

def evaluate_classifier(classifier, x_test, y_test):
    # Confusion Matrix
    y_pred = classifier.predict(x_test)
    cm = confusion_matrix(y_test, y_pred)
    class_names = np.unique(y_test)
    plot_confusion_matrix(cm, class_names)

    # Classification Report
    print("\nClassification Report:\n", classification_report(y_test, y_pred))

    # ROC Curve (not applicable for multi-class classification)

def evaluate_model(classifier, x_train, y_train, x_test, y_test):
    classifier.fit(x_train, y_train)
    prediction = classifier.predict(x_test)

    # Calculate metrics
    accuracy = accuracy_score(y_test, prediction)
    precision = precision_score(y_test, prediction, average='weighted')
    recall = recall_score(y_test, prediction, average='weighted')
    f1 = f1_score(y_test, prediction, average='weighted')
    
    print("Training Accuracy: {:.2%}".format(classifier.score(x_train, y_train)))
    print("Test Accuracy: {:.2%}".format(accuracy))
    print("Precision: {:.2%}".format(precision))
    print("Recall: {:.2%}".format(recall))
    print("F1 Score: {:.2%}".format(f1))
    

    # Evaluation
    evaluate_classifier(classifier, x_test, y_test)
    
```

### <div style="color:white;display:fill;border-radius:5px;background-color:#000000;padding:10px;letter-spacing:0.5px;overflow:hidden"><p style="padding:20px;color:white;overflow:hidden;margin:0;font-size:110%"><b></b> 7.1 | Metrics Comparison</p></div>

```python
from 
rf = RandomForestClassifier()
mlp = MLPClassifier(max_iter=1000)
lr = LogisticRegression()
knn = KNeighborsClassifier()
svm = SVC()
naive_bayes = GaussianNB()

models = [rf, mlp, lr, knn, svm, naive_bayes]
# Define evaluation metrics
metrics = {
    'Test Accuracy': [],
    'Train Accuracy': [],
    'Precision': [],
    'Recall': [],
    'F1 Score': [],
    'Cross-Validation Mean Score': [],
    'Cross-Validation Scores': []
}

# Calculate metrics for each model
for model in models:
    # Test accuracy
    test_accuracy = accuracy_score(y_test, model.predict(x_test))
    metrics['Test Accuracy'].append(test_accuracy)
    
    # Train accuracy
    train_accuracy = accuracy_score(y_train, model.predict(x_train))
    metrics['Train Accuracy'].append(train_accuracy)
    
    # Precision
    precision = precision_score(y_test, model.predict(x_test), average='weighted')
    metrics['Precision'].append(precision)
    
    # Recall
    recall = recall_score(y_test, model.predict(x_test), average='weighted')
    metrics['Recall'].append(recall)
    
    # F1 score
    f1 = f1_score(y_test, model.predict(x_test), average='weighted')
    metrics['F1 Score'].append(f1)
    
    # Cross-validation scores
    cv_scores = cross_val_score(model, x_train, y_train, cv=5, scoring='accuracy')
    metrics['Cross-Validation Scores'].append(cv_scores.tolist())
    
    # Cross-validation mean score
    cv_mean_score = np.mean(cv_scores)
    metrics['Cross-Validation Mean Score'].append(cv_mean_score)

# Create a DataFrame from the metrics
metrics_df = pd.DataFrame(metrics, index=[model.__class__.__name__ for model in models])

# Display the DataFrame
print("\nMetrics for Each Model:")
metrics_df
```

### <div style="color:white;display:fill;border-radius:5px;background-color:#000000;padding:10px;letter-spacing:0.5px;overflow:hidden"><p style="padding:20px;color:white;overflow:hidden;margin:0;font-size:110%"><b></b> 7.2 | Metrics Heatmap</p></div>

```python
# Calculate summary statistics for cross-validation scores
cv_mean_scores = metrics_df['Cross-Validation Scores'].apply(np.mean)
cv_std_scores = metrics_df['Cross-Validation Scores'].apply(np.std)

# Create a new DataFrame with summary statistics
cv_summary_df = pd.DataFrame({'Cross-Validation Mean Score': cv_mean_scores, 'Cross-Validation Std Score': cv_std_scores})

# Combine the summary statistics with other metrics
combined_metrics_df = pd.concat([metrics_df.drop(columns=['Cross-Validation Scores']), cv_summary_df], axis=1)

# Plot a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(combined_metrics_df, annot=True, cmap='coolwarm', fmt='.4f', cbar=True)
plt.title('Comparison of Models')
plt.xlabel('Metrics')
plt.ylabel('Models')
plt.xticks()
plt.tight_layout()
plt.show()
```

### <div style="color:white;display:fill;border-radius:5px;background-color:#000000;padding:10px;letter-spacing:0.5px;overflow:hidden"><p style="padding:20px;color:white;overflow:hidden;margin:0;font-size:110%"><b></b> 7.3 | Finding Best Model</p></div>

```python
# Define models with best hyperparameters
models = {
    'Random Forest': rf,
    'MLP': mlp,
    'Logistic Regression': lr,
    'KNN': knn,
    'SVM': svm,
    'Naive Bayes': GaussianNB()
}

# Initialize a list to store results
results = []

# Function to evaluate models
def evaluate_model(model, x_train, y_train, x_test, y_test):
    model.fit(x_train, y_train)
    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)

    train_accuracy = accuracy_score(y_train, y_pred_train)
    test_accuracy = accuracy_score(y_test, y_pred_test)

    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    cv_scores = cross_val_score(model, x_train, y_train, cv=cv, scoring='accuracy')

    return {
        'Train Accuracy': train_accuracy,
        'Test Accuracy': test_accuracy,
        'CV Mean Score': np.mean(cv_scores),
        'CV Std Dev': np.std(cv_scores)
    }

# Evaluate each model and store results
for model_name, model in models.items():
    metrics = evaluate_model(model, x_train, y_train, x_test, y_test)
    metrics['Model'] = model_name
    results.append(metrics)

# Create a DataFrame from the results
metrics_df = pd.DataFrame(results)

# Display results
print(metrics_df)

# Determine the best model based on the highest test accuracy
best_model_name = metrics_df.loc[metrics_df['Test Accuracy'].idxmax()]['Model']
print(f"\nBest Model: {best_model_name}\n")
```
